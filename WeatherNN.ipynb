```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch
```

```python
df=pd.read_csv("weatherdata.csv")
df.head()
```

```python
df=df.drop(['tmin','tmax'],axis=1)
df.head()
```

```python
X=df[['prcp','wspd','pres']]
y=df['tavg']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
scaler=StandardScaler()
Xtrain=scaler.fit_transform(Xtrain)
Xtest=scaler.fit_transform(Xtest)
```

```python
ytrain=np.array(ytrain)
ytest=np.array(ytest)
```

```python
Xtrain=torch.tensor(Xtrain)
Xtest=torch.tensor(Xtest)
ytrain=torch.tensor(ytrain).view(-1,1)
ytest=torch.tensor(ytest).view(-1,1)
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
#model definition

model=nn.Sequential(
    nn.Linear(3,64),
    nn.ReLU(),
    nn.Linear(64,64),
    nn.ReLU(),
    nn.Linear(64,1)
)

criterion=nn.MSELoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
#training loop

n=1000
model.train()
for epoch in range(n):
    optimizer.zero_grad()
    output=model(Xtrain)
    loss=criterion(output,ytrain)
    loss.backward()
    optimizer.step()
    print("iteration :",epoch,"loss :",loss.item())
```

```python
#test loop

model.eval()
with torch.no_grad():
    prediction=model(Xtest)
    testloss=criterion(prediction,ytest)
    print("Loss :",testloss.item())
```
